# Perform Imputation for Missing Values and  Feature Scaling and Encoding Categorical Variables using the colmn transformer technique.

we will be using the missing indicator for method to impute the missing values in the dataset. This method creates a new column for each feature with missing values and sets its value to 1 if the corresponding feature has a missing value and 0 otherwise. Then we can use this information to predict the missing values.
For feature scaling we will be using the standard scaler technique. This technique scales the features by subtracting their mean and dividing by their standard deviation. This ensures that all features are on the same scale and have similar variance.
For encoding categorical variables we will be using the ordinal encoder and one hot encoding technique. This technique assigns a numerical value to each category in a categorical variable. The categories are ordered according to some criteria such as alphabetical order or frequency of occurrence.
We will be using the column transformer technique to perform these operations on different types of features in our dataset. We will create separate transformers for numeric and categorical features and then combine them into a single column transformer object

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder

# numeric pipeline impute+scale
numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median', add_indicator=True)),
    ('scaler', StandardScaler()),
])
# categorical pipeline impute + oridinal
category_ordinal_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='Unknown')),
    ('encoder', OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1)),
])
# categorical pipeline impute + onehot
category_onehot_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='Unknown')),
    ('encoder', OneHotEncoder(handle_unknown='ignore')),
])

num_features = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_features = X_train.select_dtypes(include=['object']).columns.tolist()
stratergies_cols = categorize_encoding_strategy(X_train)
print(set(num_features) & set(stratergies_cols['ordinal_target']))
print(set(num_features) & set(stratergies_cols['one_hot']))
print(set(stratergies_cols['ordinal_target']) & set(stratergies_cols['one_hot']))

# define column transformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_pipeline, num_features),
        ('cat_ord_encoder', category_ordinal_pipeline, stratergies_cols['ordinal_target']),
        ('cat_one_encoder', category_onehot_pipeline, stratergies_cols['one_hot']),
    ]
)

# apply fit transform to training set
# X_train_imputed = preprocessor.fit_transform(X_train)
# # apply transform to testing set
# X_test_imputed = preprocessor.transform(X_test)
```

### Model Training and Evaluation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from xgboost import XGBClassifier

# --------------------------
# 3. Build model pipelines
# --------------------------

# Logistic Regression
logreg_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", LogisticRegression(max_iter=1000, class_weight="balanced"))
])

# Random Forest
rf_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", RandomForestClassifier(n_estimators=20, random_state=42, class_weight="balanced"))
])

# XGBoost
xgb_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", XGBClassifier(
        n_estimators=300,
        learning_rate=0.1,
        max_depth=6,
        subsample=0.8,
        colsample_bytree=0.8,
        use_label_encoder=False,
        eval_metric="logloss",
        scale_pos_weight= (y_train.value_counts()[0] / y_train.value_counts()[1]) # handle imbalance
    ))
])

# --------------------------
# 4. Train & evaluate
# --------------------------

pipelines = {
    "Logistic Regression": logreg_pipeline,
    "Random Forest": rf_pipeline,
    "XGBoost": xgb_pipeline
}
# load the models if it is there
model_keys = [key for key in pipelines.keys()]
trained_models = load_selective_models(model_keys)
if (len(model_keys) == len(trained_models)):
    print(f"All models loaded successfully. {trained_models.keys()}")
else:
    y_train = y_train.to_numpy().ravel()
    y_test = y_test.to_numpy().ravel()
    y_pred = {
        "Logistic Regression": None,
        "Random Forest": None,
        "XGBoost": None
    }
    # X_test_imputed = pd.DataFrame(X_test_imputed, columns=X_test.columns)
    # X_train_imputed = pd.DataFrame(X_train, columns=X_train.columns)
    for name, pipe in pipelines.items():
        pipe.fit(X_train, y_train)
        # score = pipe.score(X_test, y_test)
        y_pred[name] = pipe.predict(X_test)
        print(f"{name} Confusion Matrix: {confusion_matrix(y_test, y_pred[name])}")
        display(f"{name} Classification Report: {classification_report(y_test, y_pred[name])}")
        print(f"{name} Accuracy: {accuracy_score(y_test, y_pred[name])}")
```
